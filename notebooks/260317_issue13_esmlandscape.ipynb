#### UMAP Process

I UMAP-ed the embeddings using `umap-learn 0.5.11`, and plotted them using pyplot from the `matplotlib 3.10.8` package. The whole process can be seen in `umap_viz.ipynb`.

#### Data Preparation

I also created the table **families.csv**, which joins the families with their representative sequence, x and y UMAP coordinates, as well as size (number of members in family). This was done with the **representatives.csv** file, the UMAP data, and data pulled from petadex using `psycopg2-binary 2.9.11`.

### Results & Discussion

The UMAP and visualization went smoothly, and produced our first look at the ESM Landscape, saved in **/esm-landscape/plots**. The embeddings *appear* to form clusters, but we cannot be sure until we run controls and compare the results. Adjusting bubble size by family size looked messy on a static plot.
