### Processing Steps

Use the command 'RNAfold' or 'RNALfold' (for long RNAs, outputs local secondary structure) with the following format to read in sequence files:

```bash
RNALfold<"INPUT FILE NAME"> "OUTPUT FILE NAME".txt
```
The command outputs all local 2D structure as a string of text in the form of brackets and periods in a txt file, along with min. free energy and starting location of the hairpin as a number.  
Example:  
.(((.(.(((((((((((.((....)).))))..))))).)).).))). ( -6.40)  274

```

````python
import re
import json
with open("C:/Users/clair/RNAfold/examples/files/output5.txt", 'r') as f:
    lines = f.read().splitlines()

name = None
i = 0
results_start = {}
while i < len(lines):
    #linestrip = lines[i].strip()
    #print("start")
    if lines[i].startswith('>'):
        name = lines[i]
        results_start[name] = []
    elif(lines[i].startswith('(') or lines[i].startswith('.')) and name is not None:
            #hairpin_location = lines[i].split()[3]
            hairpin_length = len(lines[i].split()[0])
            results_start[name].append([x.strip() for x in re.findall(r'\s\d{1,3}', lines[i])])
    #print(i)
    i += 1

with open('final_output.txt', 'w') as f:
    for key, value in results_start.items():
        f.write(f"{key}: {json.dumps(value)}\n")
````
